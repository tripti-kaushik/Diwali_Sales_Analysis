```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace= True)
df 
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace= True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype(int) 
```

```python
df['Amount'].dtype
```

```python
# Exploratory Data Analysis
```

```python
df.columns
```

```python
# plotting a bar chart for Gender and it's count
ax= sns.countplot (x= 'Gender', data= df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## plotting a bar chart for gender vs total amount
sales_gen = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
sns.barplot(x= 'Gender' , y= 'Amount', data= sales_gen)
```

```python
#From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
#Age
af= sns.countplot(x='Age Group', hue='Gender',data=df)
for bars in af.containers:
    af.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group','Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sales_age
sns.barplot(x='Age Group',y='Amount', data=sales_age,hue='Gender')
```

```python
#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y='Orders', data= sales_state)
plt.title('No of Orders from Top 10 State',fontsize=20)
```

```python
# from above graph Uttar Pardesh has most order.
```

```python
#total amount/sales from top 10 state
ax= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount', data=ax,palette='Set1')
plt.title('No of Amount Sales from Top 10 State',fontsize=20)
```

```python
# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
# Marital Status
ms= sns.countplot(x='Marital_Status',hue='Gender',data=df)
for bars in ms.containers:
    ms.bar_label(bars)
```

```python
ms= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', data= ms,hue='Gender')
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
#Occupation
occ= sns.countplot(x='Occupation',data=df,palette='Set1')
sns.set(rc={'figure.figsize':(20,5)})
for bars in occ.containers:
    occ.bar_label(bars)
```

```python
occ= df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=occ,palette='Set1')
```

```python
# From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
# Product Category
pc= sns.countplot(x='Product_Category',data=df,palette='Set1')
sns.set(rc={'figure.figsize':(40,10)})
for bars in pc.containers:
    pc.bar_label(bars)
```

```python
pc= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(x='Product_Category',y='Amount',data=pc)
```

```python
#From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
# Conclusion------
# Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
```
